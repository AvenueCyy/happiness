## 赛题解读
- 本次比赛是阿里天池上的新人赛《快来一起挖掘幸福感！》,利用问卷调查所得结果分析幸福感与哪些变量有关，并进行幸福感预测。(以下摘自官网说明)https://tianchi.aliyun.com/competition/entrance/231702/information
- 赛题背景：幸福感与大家生活息息相关，每个人对幸福感都有自己的衡量标准。如果能找到影响幸福感的政策因素，便能优化资源配置来提升国民的幸福感。
- 赛题说明：赛题使用问卷调查结果，选取其中多组变量，包括个体变量（性别、年龄、地域等）、家庭变量（父母、配偶、子女、家庭资本等）、社会态度（公平、信用、公共服务等），来预测其对幸福感的评价。
- 数据说明：考虑到变量个数较多，部分变量间关系复杂，数据分为完整版和精简版两类。
- 数据来源：赛题使用的数据来自中国人民大学中国调查与数据中心主持之《中国综合社会调查（CGSS）》项目。
- 评测指标：score= ${1\over n}\sum^n_1(y_i - y^*)^2$

## 分析思路
- 1.赛题解读:
    - 鉴于评测指标采用平方和的形式，因此本次预测的结果以回归为主。
    - 本次分析采用精简版数据。
- 2.探索性数据分析与数据预处理:
    - 先对数据预处理，对其中的异常值进行填充/删除，连续数据分段处理，离散数据中分类较多的，根据目标特征进行映射处理。
    - 分析特征数据分布(结构)，及与目标变量和其它特征的关系。
- 3.特征工程
    - 特征选择：采用基于随机森林的嵌入法对特征进行筛选。删除缺失值较多或不重要的特征。
    - 特征变换：连续型数据标准化处理，对于偏态数据进行对指化处理。定序数据标签化，定类数据采用独热编码等。
    - 特征衍生：根据各个特征的关系进行人为归纳。
- 4.建模评分
    - 本次输出是回归结果，因此可先选择几个回归模型进行预测。
    - 挑选预测较好的模型进行调参优化，得出最佳模型。

```python
# 找出特征中的负值
cols = list(train_data.columns)
del cols[cols.index('survey_time')]
def negative_number(data):
    names=[]
    for name in cols:
        if name in data.columns:
            for value in data[name]:
                if value<0:
                    names.append(name)
                else:
                    pass
        else:
            pass
    d={}
    for i in names:
        if names.count(i) >= 1:
            d[i] = names.count(i)
    return d
```

```python
d1=pd.DataFrame.from_dict(negative_number(train_data),orient='index')
d1=d1.reset_index().rename(columns={'index':'feature'})
d2=pd.DataFrame.from_dict(negative_number(test_data),orient='index')
d2=d2.reset_index().rename(columns={'index':'feature'})
```

```python
c=pd.merge(d1,d2,on='feature',how='left')
c.fillna(0,inplace=True)
c.rename(columns={'0_x':'train_data','0_y':'test_data'},inplace=True)
fig=plt.figure(figsize=(15,8))
c.plot(kind='bar',x='feature',figsize=(12,3),rot=60)
plt.show()
```

- 图表展示了特征值为负值各个特征的情况，其中inc_ability的缺失值最多。目标变量中也存在异常值。
- 对于目标变量中的负值，直接将其删除。离散型特征的负值用中位数替代，连续型则用0替代。至此异常值中负值的情况处理完成。

```python
for name in cols:
    train_data.loc[train_data[name]<0,name]=np.nan
train_data.dropna(subset=['happiness'],inplace=True)
train_data[['income','family_income','family_m','house']].fillna(0)
train_data.fillna(train_data.median(),inplace=True)
```

```python
del cols[cols.index('happiness')]
for name in cols:
    test_data.loc[test_data[name]<0,name]=np.nan
test_data[['income','family_income','family_m','house']].fillna(0)
test_data.fillna(test_data.median(),inplace=True)
```

### age

- 根据survey_type和birth这两个特征，可以求出被调查者的年龄。
- 展示训练集、测试集调查人群的年龄分布，45岁~65岁的调查者居多。训练集、测试集年龄分布相近。

```python
def age(data):
    data['survey_time'] = pd.to_datetime(data['survey_time'])
    data.survey_time = pd.DatetimeIndex(data.survey_time).year
    data['age']=data.survey_time-data.birth
    return data
train_data=age(train_data)
test_data=age(test_data)
```

```python
cutPoint=[0,20,40,60,80,100]
groupLabel = [1,2,3,4,5]
train_data['age_'] = pd.cut(train_data.age,cutPoint,labels=groupLabel)
test_data['age_'] = pd.cut(test_data.age,cutPoint,labels=groupLabel)
```

```python
# 定序数据，对重要特征进行相乘处理。
Ordered_data_feature=['health', 'health_problem', 'depression','relax','edu','marital','family_status',
                                       'class','equity','status_peer','view','inc_ability','status_3_before']
def Ordered_data_feature_(data,features=Ordered_data_feature):
    for i in features:
        for j in features:
            if i != j:
                data[i+'*'+j]=data[i]*data[j]
            else:
                pass
    return data
```

```python
train_data=Ordered_data_feature_(train_data)
test_data=Ordered_data_feature_(test_data)
```

```python
def drop_columns_(data=train_data):
    drop_columns_3=[]
    for drop_columns_1 in data.columns[47:]:
        for drop_columns_2 in data.columns[47:]:
            if data[drop_columns_1].mean() == data[drop_columns_2].mean():
                drop_columns_3.append(drop_columns_1)
            else:
                pass
    drop_columns_4=list(set(drop_columns_3))
    drop_columns_4.sort(key=drop_columns_3.index)
    m=[]
    i=12
    j=0
    for k in range(1,13):
        for n in range(1,k+1):
            j=i*k+n
            m.append(j)
    drop_columns_s=[]
    for z in m:
        drop_columns_s.append(drop_columns_4[z-1])
    return drop_columns_s
```

```python
drop_columns=drop_columns_()
```

```python
# 数值型数据，对特征进行加减乘除处理。
def Numerical_data_feature_(data):
    data['family_income']+=1#family_income存在0,+1以免出现Null值。
    data['income/family_income']=data.income/data.family_income# 个人收入/家庭收入的占比
    data['area/m']=data.floor_area/data.family_m# 房屋人均面积
    data['family_income/m']=data.family_income/data.family_m# 家庭人均收入
    data['house/m']=data.house/data.family_m# 人均房产数量
    data['height_weight']=data.height_cm/data.weight_jin# 身高/体重比例
    return data
```

```python
def data_deal(data):
    data.income=np.log1p(data.income.values)
    data.floor_area=np.log1p(data.floor_area.values)
    data.family_income=np.log1p(data.family_income.values)
    columns=['income','floor_area','height_cm', 'weight_jin','family_income','family_m', 'age',
                     'income/family_income', 'area/m','family_income/m', 'house/m', 'height_weight']
    for i in range(len(columns)):
        data[columns[i]] = MinMaxScaler().fit_transform(data[columns[i]].values.reshape(-1,1)).reshape(1,-1)[0]
        warnings.filterwarnings('ignore')
    return data
```

### 绘制学习曲线

```python
def plot_learning_curve(estimator,title,X,y,ylim=None,cv=5,n_jobs=4,train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("mean_squared_error")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, scoring="neg_mean_squared_error",
        n_jobs=n_jobs, train_sizes=train_sizes)
    
    train_scores_mean = -np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

## 总结
- **1.预测结果：**
    - **模型调整**：绘制学习曲线后，发现有严重的过拟合现象。因此需对特征和模型参数进行调整，或换用其它模型进行预测对比，以求最优解。
    - **数据源**：本次分析采用精简版数据，数据源的删减会对结果产生一定影响。熟悉数据后，可采用完整数据再次进行分析预测。
- **2.分析结果(简要分析影响幸福感的重要因素)：**
    - **个体变量**：个人资产(收入，房产，户口等)对幸福感影响较小。资产多少仅表示个人活法不同，而要提升幸福感，则需要多关注身心健康，教育等。
    - **家庭变量**：婚姻，家境等是影响幸福感的重要因素。家境好的普遍幸福感也较高，另外婚姻破裂的家庭幸福感极低。
    - **社会态度**：所处阶级，公平程度，观点认可度等对幸福感影响较大，而社会态度中的诸多因素相比个体和家庭则更具影响力。说明良好稳定的社会条件才是幸福感的基石，而在动荡社会(如小说《茶馆》)中，个体和家庭经营的再好也会风雨飘摇。

## 赛题解读
- 本次比赛是阿里天池上的新人赛《快来一起挖掘幸福感！》，利用问卷调查所得结果分析幸福感与哪些变量有关，并进行幸福感预测。(以下文字摘自官网说明)https://tianchi.aliyun.com/competition/entrance/231702/information
- 赛题背景：幸福感与大家生活息息相关，每个人对幸福感都有自己的衡量标准。如果能找到影响幸福感的政策因素，便能优化资源配置来提升国民的幸福感。
- 赛题说明：赛题使用问卷调查结果，选取其中多组变量，包括个体变量（性别、年龄、地域、职业、健康、婚姻与政治面貌等等）、家庭变量（父母、配偶、子女、家庭资本等等）、社会态度（公平、信用、公共服务等等），来预测其对幸福感的评价。
- 数据说明：考虑到变量个数较多，部分变量间关系复杂，数据分为完整版和精简版两类。
- 数据来源：赛题使用的数据来自中国人民大学中国调查与数据中心主持之《中国综合社会调查（CGSS）》项目。
- 评测指标：score= ${1\over n}\sum^n_1(y_i - y^*)^2$

## 分析思路：
- 1.赛题解读:
    - 介于评测指标采用平方和的形式，因此本次预测的结果以回归为主。
    - 本次分析采用精简版数据。
- 2.探索性数据分析与数据预处理:
    - 先对数据预处理，对其中的异常值进行填充/删除，连续数据可进行分段处理。
    - 将特征按类型分为连续型变量和离散型变量，分别分析特征与目标特征，及其它相关特征的关系和本身数据结构(分布)。
- 3.特征工程
    - 特征选择：采用相关系数对特征进行筛选。对于缺失值较大且不重要的特征直接删除。
    - 特征变换：连续型数据归一化/标准化处理，有偏态，对指化处理。定序数据标签化，定类数据采用独热编码。
    - 特征衍生：根据各个特征间的关系进行人为归纳。
- 4.建模评分
    - 由于本次输出的是回归结果，因此可先选择几个回归模型进行预测。对最好的模型进行调参优化，得出相对较好的预测模型。

- province,edu,health,health_problem,depression,socialize,relax,equity,class,marital,status_peer,status_3_before,view.

- 训练集中work_status,work_yr,work_type,work_manage,丢失数据量过多，通过热力图可知，这些特征与目标量相关性较差，因此可舍弃。family_income存在1个缺失值。
- 数据类型为object的只有survey_type.其它数据为int64或float64.

```python
# 查找出各个特征值为负的个数。
cols = list(train_data.columns)
del cols[cols.index('survey_time')]
def negative_number(data):
    names=[]
    for name in cols:
        if name in data.columns:
            for value in data[name]:
                if value<0:
                    names.append(name)
                else:
                    pass
        else:
            pass
    d={}
    for i in names:
        if names.count(i) >= 1:
            d[i] = names.count(i)
    return d
```

```python
d1=pd.DataFrame.from_dict(negative_number(train_data),orient='index')
d1=d1.reset_index().rename(columns={'index':'feature'})
d2=pd.DataFrame.from_dict(negative_number(test_data),orient='index')
d2=d2.reset_index().rename(columns={'index':'feature'})
```

```python
c=pd.merge(d1,d2,on='feature',how='left')
c.fillna(0,inplace=True)
c.rename(columns={'0_x':'train_data','0_y':'test_data'},inplace=True)
fig=plt.figure(figsize=(15,8))
c.plot(kind='bar',x='feature',figsize=(12,3),rot=45)
plt.show()
```

- 图表展示了特征值为负值各个特征的情况，其中inc_ability的缺失值最多。目标变量中也存在异常值。
- 对于目标变量中的负值，直接将其删除。离散型特征的负值用中位数替代，连续型则用0替代。至此异常值中负值的情况处理完成。

```python
for name in cols:
    train_data.loc[train_data[name]<0,name]=np.nan
train_data.dropna(subset=['happiness'],inplace=True)
train_data[['income','family_income','family_m','house']].fillna(0)
train_data.fillna(train_data.median(),inplace=True)
```

```python
del cols[cols.index('happiness')]
for name in cols:
    test_data.loc[test_data[name]<0,name]=np.nan
test_data[['income','family_income','family_m','house']].fillna(0)
test_data.fillna(test_data.median(),inplace=True)
```

- 根据四分位数，对异常值进行检测。

```python
def get_error_index(all_data,k=1.5):
        q_low=all_data.quantile(q=0.25)
        q_high = all_data.quantile(q=0.75)
        q_interval = q_high - q_low
        
        q_high_=q_high+q_interval*k
        q_low_=q_low - q_interval*k
        return q_low_,q_high_
```

```python
q_low_train,q_high_train = get_error_index(train_data.floor_area)
q_low_train,q_high_train
```

```python
q_low_test,q_high_test = get_error_index(test_data.floor_area)
q_low_train,q_high_train
```

```python
def drop_feature(data):
    data.drop(columns=['city','county'],inplace=True)
    data.drop(columns=['survey_time','birth'],inplace=True)
    data.income=np.log1p(train_data.income.values)
    data.floor_area=np.log1p(train_data.floor_area.values)
    data.family_income=np.log1p(train_data.family_income.values)
    data.drop(columns=['work_status','work_yr','work_type','work_manage'],inplace=True)
    return data
```

### 特征选择

- 先建立一些特征，然后根据相关性进行筛选。
    - 前面有处理过一些特征，用同样的思路对其它特征进行融合衍生。

```python
def feature_dropfeature(data):
    data['family_income']+=1#family_income存在0,+1以免出现Null值。
    data['income/family_income']=data.income/data.family_income# 个人收入/家庭收入的占比
    data['area/m']=data.floor_area/data.family_m# 房屋人均面积
    data['family_income/m']=data.family_income/data.family_m# 家庭人均收入
    data['house/m']=data.house/data.family_m# 人均房产数量
    data['health*health_problem']=data.health*data.health_problem# 健康*疾病
    data['relax*depression']=data.relax*data.depression# 休息*沮丧
    data['height_weight']=data.height_cm/data.weight_jin# 身高/体重比例
    data['equity*class']=data['equity']*data['class']#平等*阶级
    data['house*car']=data.house*data.car#房产数量*汽车
    data['status_peer*status_3_before']=data.status_peer*data.status_3_before#与同龄人相比*与三年前相比
    data['view*class']=data['view']*data['class']#*平等
    
    data.drop(columns=['city','county'],inplace=True)
    data.drop(columns=['survey_time','birth'],inplace=True)
    data.income=np.log1p(data.income.values)
    data.floor_area=np.log1p(data.floor_area.values)
    data.family_income=np.log1p(data.family_income.values)
    data.drop(columns=['work_status','work_yr','work_type','work_manage'],inplace=True)
    return data
```

```python
train_data=feature_dropfeature(train_data)
test_data=feature_dropfeature(test_data)
```
